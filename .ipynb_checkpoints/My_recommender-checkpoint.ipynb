```python
import numpy as np
import pandas as pd
```

```python
movies_df = pd.read_csv("movies_metadata.csv")
credits_df = pd.read_csv("credits.csv")
keywords_df = pd.read_csv("keywords.csv")
```

```python
keywords_df["id"] = keywords_df["id"].astype(str)
movies_df["id"] = movies_df["id"].astype(str)
movies_df = movies_df.merge(keywords_df, on="id")
```

```python
credits_df["id"] = credits_df["id"].astype(str)
movies_df = movies_df.merge(credits_df, on="id")
movies_df
```

```python
movies_df.isna().sum()
```

```python
movies = movies_df[["id","genres","overview","title","keywords","cast_x","crew_x"]]
numerical_df=movies_df[["id","popularity","vote_average","vote_count"]]
```

```python
movies.dropna(inplace=True)
```

```python
#checking for duplicated data
movies.duplicated().sum()
movies = movies.drop_duplicates()
```

```python
#getting genres as a list

import ast
def giveList(obj):
    result = []
    for i in ast.literal_eval(obj):
        result.append(i["name"])
    return result
```

```python
movies["keywords"] = movies["keywords"].apply(giveList)
movies["genres"] = movies["genres"].apply(giveList)
```

```python
movies
```

```python
#to get the actors list

def getActorList(obj):
    result = []
    count = 0
    for i in ast.literal_eval(obj):
        if(count!=3):
            result.append(i["name"])
    return result
```

```python
movies["cast"] = movies["cast_x"].apply(getActorList)
movies.drop(columns=["cast_x"],inplace=True)
movies
```

```python
#get director name
def getDirectorName(obj):
    result = []
    for i in ast.literal_eval(obj):
        if(i["job"] == "Director"):
            result.append(i["name"])
    return result
```

```python
movies["director"] = movies["crew_x"].apply(getDirectorName)
movies.drop(columns=["crew_x"], inplace=True)
movies
```

```python
movies["genres"] = movies["genres"].apply(lambda x:[i.replace(" ","") for i in x])
movies["keywords"] = movies["keywords"].apply(lambda x:[i.replace(" ","") for i in x])
movies["cast"] = movies["cast"].apply(lambda x:[i.replace(" ","") for i in x])
movies["director"] = movies["director"].apply(lambda x:[i.replace(" ","") for i in x])
movies["overview"] = movies["overview"].apply(lambda x:x.split())
```

```python
movies
```

```python
final_df = movies[["id"]]
```

```python
final_df["tags"] = (
    movies["genres"].apply(lambda x: x if isinstance(x, list) else [x]) +
    movies["overview"].apply(lambda x: x if isinstance(x, list) else [x]) +
    movies["title"].apply(lambda x: x if isinstance(x, list) else [x]) +
    movies["keywords"].apply(lambda x: x if isinstance(x, list) else [x]) +
    movies["cast"].apply(lambda x: x if isinstance(x, list) else [x]) +
    movies["director"].apply(lambda x: x if isinstance(x, list) else [x])
)
```

```python
final_df
```

```python
final_df["tags"] = final_df["tags"].apply(lambda x: " ".join(x)) 
```

```python
final_df
```

```python
final_df["title"] = movies["title"]
```

```python
final_df.head()
```

```python
final_df["tags"] = final_df["tags"].apply(lambda x : x.lower())
```

```python
final_df["tags"][0]
```

```python
#stemming
import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()

def stem(text):
    arr = []
    for i in text.split():
        arr.append(ps.stem(i))

    return " ".join(arr)
```

```python
final_df["tags"] = final_df["tags"].apply(stem)
final_df
```

```python
final_df["tags"][0]
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf_vectorizer = TfidfVectorizer(stop_words='english',max_features=5000) 
tfidf_matrix = tfidf_vectorizer.fit_transform(final_df["tags"])
```

```python
# now calculating the similarity between movies using cosine similarity
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(tfidf_matrix)
```

```python
def recommend(movie):
    if movie not in final_df["title"].values:
        print("Movie not found in dataset.")
        return
    
    # Find the index of the movie
    movie_index = final_df[final_df["title"] == movie].index[0]

    # Get similarity scores for the movie
    distance_arr = similarity[movie_index]

    # Get top 5 most similar movies (excluding the movie itself)
    movies_list = sorted(
        list(enumerate(distance_arr)), 
        reverse=True, 
        key=lambda x: x[1]
    )[1:6]  # Skip the first one (itself)

    # Print recommended movies and genres
    for i in movies_list:
        print(movies.iloc[i[0]].title, "-", movies.iloc[i[0]].genres)
```

```python
recommend("Jumanji")
```
